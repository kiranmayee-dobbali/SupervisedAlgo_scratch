```python
from scipy.sparse import csr_matrix
import numpy as np
```

```python

def read_libsvm(fname, num_features=0):
    '''
        Reads a libsvm formatted data and outputs the training set (sparse matrix)[1], 
        the label set and the number of features. The number of features
        can either be provided as a parameter or inferred from the data.

		Example usage:
		
		X_train, y_train, num_features = read_libsvm('data_train')
		X_test, y_test, _ = read_libsvm('data_test', num_features)

		[1] https://docs.scipy.org/doc/scipy/reference/generated/scipy.sparse.csr_matrix.html
	'''
    data = []
    y = []
    row_ind = []
    col_ind = []
    with open(fname) as f:
        lines = f.readlines()
        for i, line in enumerate(lines):
            elements = line.split()
            y.append(int(elements[0]))
            for el in elements[1:]:
                row_ind.append(i)
                c, v = el.split(":")
                col_ind.append(int(c))
                data.append(float(v))
    if num_features == 0:
        num_features = max(col_ind) + 1
    X = csr_matrix((data, (row_ind, col_ind)), shape=(len(y), num_features+1))

    return X, np.array(y), num_features
```

```python
def predict(row, w , b ):
    mul=(np.dot(np.array(row),w)+b)
    result = -1
    if mul>=0:
        result = 1
    return result

def update(row_df, y, w, b, lr):
    w_new=[]
    b_new = 0
    for i, ele in enumerate(row_df):
        w_new.append(w[i]+lr*y*ele)
        b_new = b + lr*y
    w_new = np.array(w_new)   
    return w_new , b_new    

def perceptron(X_df, y_t , epochs=20, lr=0.01):
    rgen = np.random.RandomState(0)
    b = rgen.uniform(-0.01,0.01,1) #initialize bias
    w = rgen.normal(loc=0.0, scale=0.01, size=361)
    arr_acc=[]
    error=[]
    #op_predict =[]
    for e in range(epochs):
        errors=0
        #print("Epoch : {}".format(e))
        for index, row in X_df.iterrows():
            y_dash=predict(row ,w ,b)
            #op_predict.append(y_dash)
            if(y_t[index]!=y_dash):
                w, b = update(row, y_t[index], w, b, lr)
                errors=errors+1
        error.append(errors)
        #print("Accuracy in epoch : {} ".format((len(y_t)-errors)/len(y_t)))
        arr_acc.append((len(y_t)-errors)/len(y_t))
    accuracy = (len(y_t)-errors)/len(y_t)  
    return w,b,error, accuracy, arr_acc   
```
